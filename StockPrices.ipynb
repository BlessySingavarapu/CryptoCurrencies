```python
import yfinance as yf
import pandas as pd
```

```python
tickers = ["^GSPC", "^IXIC", "^NSEI"]

start_date = "2020-01-01"
end_date = "2025-09-30"
```

```python
all_data = []

for ticker in tickers:
    print(f"Downloading data for {ticker}...")
    
    df = yf.download(
        ticker,
        start=start_date,
        end=end_date,
        interval="1d",
        auto_adjust=False,
        progress=False
    )
    
    df = df.reset_index()
    df["ticker"] = ticker

    df = df[[
        "ticker", "Date", "Open", "High", "Low", "Close", "Adj Close", "Volume"
    ]]

    df.columns = [
        "ticker", "date", "open", "high", "low", "close", "adj_close", "volume"
    ]

    all_data.append(df)

df_stocks = pd.concat(all_data, ignore_index=True)
```
