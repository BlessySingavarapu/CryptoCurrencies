```python
import requests
```

```python
import time
```

```python
# Base URL for the API Call
```

```python
base_url = "https://api.coingecko.com/api/v3/coins/markets"
```

```python
# Function for the API Call to fetch the CRYPTO CURRENCY Coin MetaData

# Given URL -  https://api.coingecko.com/api/v3/coins/markets?vs_currency=inr&per_page=250&order=market_cap_desc&page=1&sparkline=False
```

```python
def fetch_all_crypto_metadata(vs_currency="inr", per_page=250):
# ALL Coins list as all_coins[]
    all_coins = []
    page = 1

# Rap the url parameters
    while True:
        parameters = {
            "vs_currency": vs_currency,
            "order": "market_cap_desc",
            "per_page": per_page,
            "page": page,
            "sparkline": False
        }
# API Call to get the data of coins
        response = requests.get(base_url, params=parameters)

# If Status Code of response from API Call is Success i.e., 200 then break the loop
        if response.status_code != 200:
            print(f"Error on page {page}: {response.status_code}")
            break

# Collect the Data which is in the form of JSON
        data = response.json()

# Stop when no more data
        if not data:
            break
            
# Push the Data into the list all_coins
        all_coins.extend(data)
        print(f"Fetched page {page} | Records: {len(data)}")

        page += 1
        time.sleep(5)  # avoid rate limiting

# Return the data of all_coins in the form of DataFrame to the frunction
    return pd.DataFrame(all_coins)
```

```python
# Fetch all crypto metadata  --   Function calling
# df_crypto is a DataFrame
df_crypto = fetch_all_crypto_metadata()

# Display basic info of df_crypto MetaData
print(df_crypto.head())
print(f"\nTotal coins fetched: {len(df_crypto)}")

# Save the DataFrame df_crypto into to CSV file as coingecko_crypto_metadata
df_crypto.to_csv("D:\\Python_Code\\Project1-Cross-MarketAnalysisCryptoOil&StocksWithSQL&Streamlit\\coingecko_crypto_metadata.csv", index=False)
```
