```python
import pandas as pd

url = "https://raw.githubusercontent.com/datasets/oil-prices/main/data/wti-daily.csv"

df_oil = pd.read_csv(url)
```

```python
df_oil["Date"] = pd.to_datetime(df_oil["Date"])
```

```python
start_date = "2020-01-01"
end_date = "2026-01-31"

df_oil_filtered = df_oil[
    (df_oil["Date"] >= start_date) &
    (df_oil["Date"] <= end_date)
]
```

```python
print(df_oil_filtered.head())
print(df_oil_filtered.tail())
print("Total records:", len(df_oil_filtered))
```

```python
# (Optional) Rename columns for clarity
df_oil_filtered = df_oil_filtered.rename(
    columns={"Price": "wti_price_usd"}
)
```

```python
df_oil_filtered.to_csv(
    "wti_oil_prices_Jan_2020_to_Jan_2026.csv",
    index=False
)
```
